# Housing Market Analysis

## imports

```python
import pandas as pd
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, HTML
```

## load data

```python
data = pd.read_csv("portland_housing.csv", index_col=0, low_memory=False)
data.head()
```

## sample data with completeness

```python
# Calculate the percentage of non-missing values for each column
percent_filled = data.notna().mean() * 100

# Get a sample non-None value for each column
sample_values = data.apply(
    lambda col: col.dropna().iloc[0] if not col.dropna().empty else "None"
)

# Create a DataFrame to display the results and sort by percentage
percent_filled_df = pd.DataFrame(
    {"Percent Filled": percent_filled, "Sample Value": sample_values}
).sort_values(by="Percent Filled", ascending=False)

# Transpose the DataFrame for horizontal display
percent_filled_transposed = percent_filled_df.T

# Display the DataFrame with pretty formatting
display(HTML(percent_filled_transposed.to_html()))
```
